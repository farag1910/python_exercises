```python
class person:
    def __init__ (self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def display_info(self):
        print(f"name is {self.name},age is {self.age}, gender is {self.gender}")
class Assistant(person):
    def __init__(self,name,age,gender,salary):
        super().__init__(name,age,gender)
        self.salary=salary
        self.courses=[]
    def add_course(self,course):
        self.courses.append(course)
        print(f"course: {course.name} is assigned to assistant :{self.name}")
    def display_info(self):
        super().display_info()
        print(f"salary is {self.salary}")
        print("courses : ")
        for course in self.courses:
            print(f"-{courses.name}")
class Student(person):
    def __init__(self,name,age,gender,gpa,specialization):
        super().__init__(name,age,gender)
        self.gpa=gpa
        self.specialization=specialization
    def display_info(self):
        super().display_info()
        print(f"gpa is {self.gpa} specilization is {self.specialization}")
class Course:
    def __init__(self,name,specialization):
        self.name=name
        self.specialization=specialization
    def display_info(self):
        print(f" name is {self.name} specialization is {self.specialization}")
class Doctor(person):
    def __init__(self,name,age,gender,salary):
        super().__init__(name,age,gender)
        self.salary=salary
        self.courses=[]
        self.assistants=[]
    def display_info(self):
        super().display_info()
        print(f"salary is {self.salary}")
        print("courses : ")
        for course in self.courses:
            print(f"-{course.name}")
        print("assistants :")
        for assistant in self.assistants:
            print(f"-{assistant.name}")
    def add_assisstant(self,assistant):
        self.assistants.append(assistant)
        print(f"assistant: {assistant.name} assigned to doctor:{self.name} " )
    def add_course(self,course):
        self.courses.append(course)
        print(f"course: {course.name} assigned to doctor:{self.name} " )  
class College:
    def __init__(self,name,location):
        self.name=name
        self.location=location
        self.doctors=[]
        self.assistants=[]
    def add_doctor(self,doctor):
        self.doctors.append(doctor)
        print(f"doctor {doctor.name} is assigned to the college")
    def add_assistant(self,assistant):
        self.assistants.append(assistant)
        print(f"assistant {assistant.name} is assigned to the college")
    def display_info(self):
        print (f"name is {self.name}, location is {self.location}")
        print(" doctors :")
        for doctor in self.doctors:
            print (f"-{doctor.name}")
        print(" assistants :")
        for assistant in self.assistants:
            print (f"-{assistant.name}")   
```
